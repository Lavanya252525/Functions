1. What is the difference between a function and a method in Python?
 - In python the key difference between a function and a method is:
 - Function is a standalone block of code that performs a task and is not assiciated with any object. A function is defined using def keyword.
 - Whereas, A method is a function that belongs to an object. It is called using dot(.) notation. It automatically takes the instance (self) as the first parameter.



```python
#For example of a function
def add(a,b):
  return a+b
add(2,3)
```

```python
#For example of a Method
class person:
  def say_hello(self):
    return "Hello!"
p = person()
print(p.say_hello())
```

6. Explain the concept of generators in Python and how they are defined.
 - Generator is a special type of function used to generate a squence of values, one at a time. They are defined using yield instead of return.

```python
def square_no_gen(n):
    for i in range(n):
        yield i**2
gen = square_no_gen(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

7. What are the advantages of using generators over regular functions?
 - Regular function calculates the square of the values and returns in one go since it is in one go, it will use more memory space and it there are a lot of elements, it could take a lot of execution time as all the computations will be done first and then will be returned one go.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - These are all functional programming tools used to process data, but each has a different purpose.
 -**map()** applies a function to every item in an iterable and returns a new iterable with transformed values.

```python
numbers = [1, 2, 3]
list(map(lambda x: x**2, numbers))
```

 - **filter()** Filters items based on a condition (True/False). It keeps only those items for which the function returns True.  

```python
l = [2, 1, 1, 3, 4, 4, 5, 6]
list(filter(lambda x:x%2==0 , l))
```

 - **reduce()** combines all items in an iterable into a single value. But it must be imported from functools in python.

```python
from functools import reduce
L = [ 1, 2, 3, 4, 5, 6]
reduce(lambda x, y:x*y, L)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
def is_prime(n):
  if n < 2:
    return False
  for i in range(2, n):
    if n % i == 0:
      return False
  return True

for number in range (1, 201):
  if is_prime(number):
    print(number)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read_lines(filename):
  for line in open(filename):
    yield line

for line in read_lines("text.file"):
  print(line)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(text):
  vowels = "aeiouAEIOU"
  filtered = filter(lambda x: x not in vowels, text)
  return filtered
remove_vowels("Lavanya Rawat")
```
